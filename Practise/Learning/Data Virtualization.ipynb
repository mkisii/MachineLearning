```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(5,100,100)
y = np.linspace(10,1000,100)
```

```python
#ploting a two 1-D numpy arrays
plt.plot(x,y)
```

```python
 plt.plot ([1,3,4,5,6,7],[2,4,6,7,8,9])
 plt.show()   
```

```python
# Axis labels and title of graph
plt.plot(x, y)

# x and y labels, and title
plt.xlabel('Current')
plt.ylabel('Voltage')
plt.title("Ohm's Law")

# Defining the range of labels of axis
plt.xlim([20,80])
plt.ylim([200,800])
plt.show()
```

```python
#Ploting of multiple lines on same plot
x = np.linspace(0, 5, 20)
y = np.linspace(4, 8, 20)

# plot three curves y , y**2 and y**3 with different line type
plt.plot(x,y, 'r-',x,y**2, 'b+', x,y**3, 'g^')
plt.show()
```

# Having a Subplot

```python
x = np.linspace(10,100,1000)
y = np.log(x)

# initiate the figure
plt.figure(1)

# creating a subplot with 1 row and 2 columns

# first subplot
plt.subplot(121)
plt.title('y = log(x)')
plt.plot(x,y)


# second subplot
plt.subplot(122)
plt.title('y = log(x)**2')
plt.plot(x,y**2)
plt.show()

```

```python
x = np.linspace(1,10,100)

plt.figure(1)
# subplot 1
plt.subplot(2,2,1)
plt.title('Linear')
plt.plot(x,x)

#subplot 2
plt.subplot(2,2,2)
plt.title('Cubic')
plt.plot(x,x**3)

plt.figure(2)
# subplot 3
plt.subplot(2,2,1)
plt.title('Log')
plt.plot(x, np.log(x))

# subplot 4
plt.subplot(2,2,2)
plt.title('Square')
plt.plot(x,x**2)
plt.show()
```

```python
x = np.linspace(10,100,1000)

plt.subplot(3,4,1)
plt.title('Linear')
plt.plot(x,x)
```

```python
# plot  a simple density
sns.distplot(data['Shipping_Cost'])
plt.show()
```

```python
plt.subplot(2,2,1)
plt.title('Sales')
sns.distplot(data['Sales'])

plt.subplot(2,2,2)
plt.title('Profit')
sns.distplot(data['Profit'])

plt.subplot(2,2,3)
sns.distplot(data['Order_Quantity'])

plt.subplot(2,2,4)
sns.distplot(data['Shipping_Cost'])
plt.show()
```

```python
# Ploting joint plots 
# profit and Sales 

sns.jointplot('Sales', 'Profit', data)
plt.show()
```

```python
# removing outlayers from the jointplot
data = data[(data.Sales < 20000) & (data.Profit < 10000)]
sns.jointplot('Sales', 'Profit', data)
plt.show()
```
