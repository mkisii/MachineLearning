Pandas Dataframe

```python
#Creating random values
temp = np.random.randint(low = 20, high = 100, size=[20,])
name = np.random.choice(['C++','Js','Java','Python'],20)
random = np.random.choice([10,11,13,14,15,16],20)
```

```python
#Creating a zip file of random generated values
a = list(zip(temp,name,random))
```

```python
#Creating a  DataFrame using a list
df =pd.DataFrame(data=a,columns=['temp','name','random'])
```

```python
df
```

```python
# Creating a DataFrame using dictionaries
df = pd.DataFrame({'temp':temp, 'name':name, 'random':random})
```
