```python
#graphs
x = np.linspace(0,5,200)
y1 =2 *x
y2 = x**2
y3 = np.log(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
#making subplots
x = np.linspace(0,5,200)
y1 =np.sin(x)
y2 = np.sqrt(x)
plt.subplot(211)
plt.plot(x,y1,'r-')
plt.subplot(212)
plt.plot(x,y2,'g--')
plt.show()
```

```python
#Having mutltiple plotting windows
plt.figure(1)
plt.plot(x,y1,'b-')
plt.figure(2)
plt.plot(x,y2,'y--')
plt.show()
```

```python
#Setting titles for graphs
x =np.linspace(0,20,200)
y = np.sin(x)
plt.title("Sin Function")
plt.suptitle("Data SCience")
plt.grid(True)
plt.plot(x,y)

#Labelling axes
plt.xlabel("x_values")
plt.ylabel("y_values")
plt.show()
```

```python
#Labeling using Legends Functions
x = np.linspace(10,50,500)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x/3)
plt.plot(x,y1,'b-', label = "Sine")
plt.plot(x,y2,'r-',label = "Cosine")
plt.plot(x,y3, 'g-',label ="Logarithm")
plt.legend(loc = "upper left")
plt.show()
# saving of diagrams
plt.savefig("functions.png")
```

#Matplotlib plot types

```python
# Histogram
mu , sigma = 172,4

x = mu + sigma * np.random.randn(10000)

plt.hist(x,100, density=True, facecolor = "blue")

plt.xlabel("Height")
plt.ylabel("Plobability")
plt.title("Height of Students")
plt.text(160,0.125, " mu = 172 , sigma = 4")
plt.axis([155,190,0,0.15])
plt.grid(True)
         
```

```python
#Bar Charts
bob = (90,87,67,78)
char = (80,80,46,66)
dan = (40,95,76,89)
skills = ("Python", "Java", "Networking", "Machine Learning")

width = 0.2
index = np.arange(4)
plt.bar(index, bob, width=width,label = "Bob")

plt.bar(index+width,char, width =width , label ="Char")

plt.bar(index + width * 2,dan,width =width, label = "Dan")

plt.xticks(index + width, skills)
plt.ylim(0,120)
plt.title("IT Skill Levels")
plt.ylabel("Skills Levels")
plt.xlabel("IT Skill")
plt.legend()
```

```python
#Pie Chart
labels = ("Kenyans", "Tanzanians","Ugandans","Others")
values = (40,35,25,18)

plt.pie(values,labels=labels, autopct = "%.2f%%", shadow=True)

plt.title("Students Nationalities")
plt.show()
```

```python
#Scatter PLots
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x,y)

plt.show()
```

```python
#Box plots
mu, sigma = 172,4
values = np.random.normal(mu,sigma,200)

plt.boxplot(values)
plt.title("Student's Height")
plt.ylabel("Height")
plt.show()
```

```python
# 3D plots
from mpl_toolkits import mplot3d
ax = plt.axes(projection = '3d')
plt.show()
```

```python
z = np.linspace(0,20,100)
x = np.sin(z)
y = np.cos(z)

ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z)
plt.show()
```

```python
#Making surface plots
ax = plt.axes(projection = '3d')
def z_function(x,y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)

x,y = np.meshgrid(x,y)
Z = z_function(x,y)

ax.plot_surface(x,y,Z)
plt.show()
```
