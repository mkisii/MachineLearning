```
import urllib
import folium
from folium import plugins
import csv
```

LINK FOR THE CSV FILE
https://drive.google.com/file/d/18cGdrANz6s9nkzMgbyr2RHWNXng-rU8F/view?usp=sharing

```
file = open("/content/drive/MyDrive/damage_assessment_NDVI.csv")
reader = csv.reader(file, delimiter=',')
title = True
lats = []
lons = []
mag_cropland = []
mag_pastureland = []
mag_forestland = []
for row in reader:
    if title:
      title = False
    else:
      lats.append(float(row[1]))
      lons.append(float(row[2]))
      mag_cropland.append(float(row[4]))
      mag_pastureland.append(float(row[5]))
      mag_forestland.append(float(row[6]))
```

```
print(lats)
print(lons)
print(mag_cropland)
print(mag_pastureland)
print(mag_forestland)
```

#CROPLAND

```
# Using USGS style tile
url_base = 'http://server.arcgisonline.com/ArcGIS/rest/services/'
service = 'NatGeo_World_Map/MapServer/tile/{z}/{y}/{x}'
tileset = url_base + service

m = folium.Map(location=[37.8716, -122.2727], zoom_start=2,\
                control_scale = True, tiles=tileset, attr='USGS style')
m.add_children(plugins.HeatMap(zip(lats, lons, mag_cropland), radius = 20))
```
