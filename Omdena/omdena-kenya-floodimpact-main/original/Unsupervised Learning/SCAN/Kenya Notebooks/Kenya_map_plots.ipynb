#PLotting the function

```

import folium
from IPython.display import display
import re
import pandas as pd 
import os

def plotOnMap(cluster_path,save_path):

  #Plot all points of a given cluster on the folium map  
  x,y = [],[]

  cluster_data = list(map(lambda x :'kenya_tiff/image_'+re.findall('\d+',x)[-1]+".tif" if 'jpeg' in x else x,os.listdir(cluster_path)))


  print('no. of images :',len(cluster_data))

  bf_tif = cluster_data[0]



  #defining a function for getting the bounds

  bf = rio.open(bf_tif)
  bottom_0 = bf.bounds[1] #first 'before' image
  right_1 = bf.bounds[2]
  
  y.append(bottom_0)
  x.append(right_1)


  m = folium.Map(location=[bottom_0,right_1],zoom_start=7)

  for bf_tif in cluster_data:

    if  'tif' in bf_tif:

        raster = rio.open(bf_tif)
        bottom = raster.bounds[1]
        right = raster.bounds[2]    
        y.append(bottom)
        x.append(right)
        folium.Marker(location=[bottom,right],popup='Default popup Marker1',tooltip='Click here to see Popup').add_to(m)
      
    else:
        print(bf_tif+" not found!")

  #Displaying the Map
  display(m)

  if not os.path.exists(save_path):
      os.mkdir(save_path)

  #saving the orgs.
  csv_name = save_path+"_"+cluster_path[:-1].split('/')[-1]+'.csv'
  pd.DataFrame({'x':x,'y':y}).to_csv(f"{save_path}/{csv_name}",index = False)
  print(csv_name,'file created.')
```
